| 1 | Sara | Davis | 1 |

| 3 | Judy | Lew | 1 |

| 5 | Sven | Mortensen | 5 |
| 6 | Paul | Suurs | 5 |
| 7 | Russell | King | 5 |
| 8 | Maria | Cameron | 5 |
| 9 | Patricia | Doyle | 5 |

(45 row(s) affected)

```sql
USE Northwinds2022TSQLV7
SELECT EmployeeId, EmployeeFirstName, EmployeeLastName, Copies.N
FROM HumanResources.Employee
CROSS JOIN dbo.Nums AS Copies
WHERE Copies.N <= 5
ORDER BY Copies.N, EmployeeId;
```

**<font color=#202f96>Explanation:</font>**  
The proposition is to return 5 copies of each row from the "Employees" table. We use the select clause to select EmployeeId, EmployeeFirstName, EmployeeLastName, and Copies.N columns. We use the from clause to indicate the involved HumanResources.Employee and dbo.Nums. We use cross join clause to combine each row from the Employees table with every row from the Copies table. The predicate is our where clause and it ensures only 5 copies of each employee are generated. We use the order by clause to set the results in ascending order by copies.n and employee id.

| 52 | Customer PZNLA | No |

| ... | ... | ... |

(91 row(s) affected)

```sql
USE Northwinds2022TSQLV7
SELECT C.CustomerID, C.CustomerCompanyName, 
    CASE WHEN O.OrderId IS NOT NULL THEN 'Yes' ELSE 'No' END AS HasOrderOn20160212
FROM Sales.Customer AS C
    LEFT OUTER JOIN Sales.[Order] AS O
    ON O.CustomerId = C.CustomerId
    AND O.OrderDate = '20160212'
```

<font color=#8439b3>**Explanation:**</font><br>
This proposition is asking is to return the customer details along with a Yes/No indication based on whether they placed an order on February 12, 2016. We will use two primary tables: Sales.Customer and Sales.[Order]. We select columns from the Sales.Customer table, named CustomerID and CustomerCompanyName. The LEFT OUTER JOIN clause is used to incorporate information from the Sales.[Order] table. The connection between the two tables is made through the CustomerId column. To focus on orders placed on February 12, 2016, we use the condition O.OrderDate = '20160212' in the ON clause. The CASE WHEN statement acts as a predicate, evaluating whether the OrderId is not null for each customer. If true, 'Yes' is assigned to the new column HasOrderOn20160212; otherwise, 'No' is assigned.
