-- Desired output
custid      companyname

(2 row(s) affected)

```sql

TSQL
USE TSQLV4
SELECT c.custid, c.CompanyName
FROM Sales.Customers c
LEFT JOIN dbo.Orders o ON c.custid = o.custid
WHERE o.OrderID IS NULL;


NORTHWINDS
USE Northwinds2022TSQLV7
SELECT c.CustomerID, c.CustomerCompanyName
FROM Sales.Customer c
LEFT JOIN Sales.[Order] o ON c.CustomerID = o.CustomerId
WHERE o.OrderId IS NULL;
```


EXPLANATION
Here the desired output was customers who have placed no orders. Left join on Sales.Customer so that no uncessary information is brought in for analysis from the order table and then ordered it where OrderId is null to get the desired output.
