## _Top 1: Exercise 6_

- Return customers with orders placed on Feb 12, 2016 along with their orders
- Tables involved: TSQLV4 database, Customers and Orders table

Desired output for Northwinds2022TSQLV7 database:

| CustomerId | CustomerCompanyName | OrderId | OrderDate |
| --- | --- | --- | --- |
| 48 | Customer DVFMB | 10883 | 2016-02-12 |
| 45 | Customer QXPPT | 10884 | 2016-02-12 |
| 76 | Customer SFOGW | 10885 | 2016-02-12 |

```sql
-- Use TSQLV4
-- Query to be converted

USE TSQLV4

SELECT  c.custid, 
        c.companyname,
        o.orderid,
        o.orderdate
FROM Sales.Customers AS c 
INNER JOIN Sales.Orders AS o 
    ON o.custid = c.custid
WHERE o.orderdate = '20160212';

-- Use Northwind2022TSQLV7
-- Converted Query

USE Northwinds2022TSQLV7

SELECT	c.CustomerId,
	    c.CustomerCompanyName,
	    o.OrderId,
	    o.OrderDate
FROM Sales.Customer AS c
INNER JOIN Sales.[Order] AS o
	ON o.CustomerId = c.CustomerId
WHERE o.OrderDate = '20160212';
```

## **Explanation:**

This proposition queries and retrieves customer and order information from a database named "Northwinds2022TSQLV7". It uses the "Northwinds2022TSQLV7" database context and selects four columns: c.CustomerId, c.CustomerCompanyName, o.OrderId, and o.OrderDate. The "Customer" table is aliased as "c" while the "Order" is aliased as "o", and an INNER JOIN is performed between the "Customer" and "Order" tables where CustomerId is the column being intersected by both tables. The "WHERE" clause filters the result set to include orders placed on February 12, 2016.

```sql
-- Use TSQLV4
-- Query to be converted

USE TSQLV4

SELECT	e.empid, 
		e.FullName, 
		COUNT(o.orderid) AS TotalOrders, 
		SUM(od.qty) AS TotalUnitSold
FROM HR.Employees AS e
LEFT JOIN Sales.Orders AS o ON e.empid = o.empid
LEFT JOIN Sales.OrderDetails AS od ON o.orderid = od.orderid
GROUP BY e.empid, e.FullName;

-- Use Northwind2022TSQLV7
-- Converted Query

USE Northwinds2022TSQLV7

SELECT	e.EmployeeId, 
		e.EmployeeFirstName,
		e.EmployeeLastName, 
		COUNT(o.OrderId) AS TotalOrders, 
		SUM(od.Quantity) AS TotalUnitSold
FROM HumanResources.Employee AS e
LEFT JOIN Sales.[Order] AS o ON e.EmployeeId = o.EmployeeId
LEFT JOIN Sales.[OrderDetail] AS od ON o.OrderId = od.OrderId
GROUP BY e.EmployeeId, e.EmployeeFirstName, e.EmployeeLastName;
```

## **Explanation:**

This code retrieves a list of employees, their total order processing and sales, by specifying columns e.empid and e.FullName. It calculates COUNT(o.orderid) and SUM(od.qty) as TotalOrders and TotalUnitSold, respectively. The data is aliased as "e" in the "HR" schema. The results are joined with the "Employees" table in the "Sales" schema, ensuring all employees are included. The results are then grouped by employee ID and full name, providing insights into their performance.
