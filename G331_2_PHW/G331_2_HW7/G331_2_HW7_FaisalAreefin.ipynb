## _TOP 3_

Proposition:

- <span style="font-size: 14px;">Write a query to calculate the total quantity of orders for each customer ('A', 'B', 'C', 'D') handled by each employee.</span>
- <span style="font-size: 14px;">Tables involved: dbo.Orders</span>

```sql
--Query against Orders, grouping by employee, pivoting customers,
-- aggregating sum of quantity
USE TSQLV6
SELECT empid,
SUM(CASE WHEN custid = 'A' THEN qty END) AS A,
SUM(CASE WHEN custid = 'B' THEN qty END) AS B,
SUM(CASE WHEN custid = 'C' THEN qty END) AS C,
SUM(CASE WHEN custid = 'D' THEN qty END) AS D  
FROM dbo.Orders
GROUP BY empid;

USE Northwinds2022TSQLV7
SELECT EmployeeId,
  SUM(CASE WHEN CustomerId = 'A' THEN Quantity END) AS A,
  SUM(CASE WHEN CustomerId = 'B' THEN Quantity END) AS B,
  SUM(CASE WHEN CustomerId = 'C' THEN Quantity END) AS C,
  SUM(CASE WHEN CustomerId = 'D' THEN Quantity END) AS D  
FROM dbo.Orders
GROUP BY EmployeeId;
```

**Explanation**

The "Orders" table is used to calculate the sum of quantities for each customer type ('A', 'B', 'C', 'D'). The "SUM(CASE WHEN custid = 'A' THEN qty END) AS A" function checks if the customer ID is 'A' and adds the quantity for that row to the total sum for customer A. The results are then grouped by empid, with the sums calculated separately for each unique empid.

- <span style="color: var(--vscode-foreground);">Tables involved: dbo.Orders</span>

```sql
USE TSQLV6;

-- Logical equivalent of previous query using the native PIVOT operator
SELECT empid, A, B, C, D
FROM (SELECT empid, custid, qty
      FROM dbo.Orders) AS D
  PIVOT(SUM(qty) FOR custid IN(A, B, C, D)) AS P;

-- Query demonstrating the problem with implicit grouping
SELECT empid, A, B, C, D
FROM dbo.Orders
  PIVOT(SUM(qty) FOR custid IN(A, B, C, D)) AS P;

use Northwinds2022TSQLV7
-- Logical equivalent of previous query using the native PIVOT operator
SELECT EmployeeId, A, B, C, D
FROM (SELECT EmployeeId, CustomerId, Quantity
      FROM dbo.Orders) AS D
  PIVOT(SUM(Quantity) FOR CustomerId IN(A, B, C, D)) AS P;

-- Query demonstrating the problem with implicit grouping
SELECT EmployeeId, A, B, C, D
FROM dbo.Orders
  PIVOT(SUM(Quantity) FOR CustomerId IN(A, B, C, D)) AS P;
```

**Explanation**

The query selects the empid column and columns representing the sum of quantities for each customer type ('A', 'B', 'C', 'D') from the "Orders" table. The columns are renamed as 'A', 'B', 'C', 'D'. A derived table (D) is created by selecting the empid, custid, and qty columns from the "Orders" table. The data is pivoted from D, calculated for each customer type ('A', 'B', 'C', 'D') and placed in separate columns. The PIVOT operation transforms the data from rows to columns based on the values in the custid column. This process is performed from the "Orders" table, which belongs to the default schema in the database.
