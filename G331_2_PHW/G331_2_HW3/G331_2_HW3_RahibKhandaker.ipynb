# Top 1 - Chapter 4 Exercise 4

- Write a query that returns employees who did not place orders on or after May 1, 2016:
- Tables involved: HR.Employees and Sales.Orders

Desired Output:

| empid | FirstName | lastname |
| --- | --- | --- |
| 3 | Judy | Lew |
| 5 | Sven | Mortensen |
| 6 | Paul | Suurs |
| 9 | Patricia | Doyle |

```sql

-- Chapter 4 Exercise 3
USE TSQLV4
SELECT empid, FirstName, lastname from HR.Employees
where empid not in
(select empid from Sales.Orders
where orderdate >= '20160501')

USE Northwinds2022TSQLV7
SELECT EmployeeId, EmployeeFirstName, EmployeeLastName from HumanResources.Employee
where EmployeeId not in
(select EmployeeId from Sales.[Order]
where orderdate >= '20160501')
```

## Top 5 - Chapter 5 Q6-The Apply Operator

- Return with OFFSET-FETCH the three most recent orders for each customer
- Tables Included: Sales.Customers, Sales.Orders

263 Rows Affected

```sql


-- **Chapter 5**

-- With OFFSET-FETCH return the three most recent orders for each customer
--use TSQLV6
--SELECT C.custid, A.orderid, A.orderdate FROM Sales.Customers AS C
--CROSS APPLY
--(SELECT orderid, empid, orderdate, requireddate FROM Sales.Orders AS O
--WHERE O.custid = C.custid
--ORDER BY orderdate DESC, orderid DESC
--OFFSET 0 ROWS FETCH NEXT 3 ROWS ONLY) AS A;

use Northwinds2022TSQLV7
SELECT C.customerid, A.orderid, A.orderdate FROM Sales.Customer AS C
CROSS APPLY
(SELECT orderid, employeeid, orderdate, requireddate FROM Sales.[Order] AS O
WHERE O.customerid = C.customerid
ORDER BY orderdate DESC, orderid DESC
OFFSET 0 ROWS FETCH NEXT 3 ROWS ONLY) AS A;

```

We use the select clause to select the columns custid from the Sales.Customers table and the orderid and orderdate from the inner query that uses the sales.orders table. Then use the cross apply operator to get the results of the inner query for each row in the left table if it is not empty. The inner query gets the top 3 rows of the orderid, empid, orderdate, and requireddate from the sales.orders table if the customerid from the orders table matches the customerid of the customers tables that is outside this query and it gets ordered by the date and the orderid to get the most recent orders. All together returns each employee's three most recent orders.</span>
