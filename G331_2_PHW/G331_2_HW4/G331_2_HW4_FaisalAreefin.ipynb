This helps find customers who placed orders in both months and employees who served them.

## _Top 3: The EXCEPT ALL Multiset Operator_

- Write a query that obtains distinct employee locations not present in customer locations. 
- Tables involved: Sales.Customers, HR.Employees

Desired output for TSQLV6 database:

| country | region | city |
| --- | --- | --- |
| USA | WA | Redmond |
| USA | WA | Tacoma |
| USA | WA | Seattle |

Desired output for Northwinds2022TSQLV7 database:

| EmployeeCountry | EmployeeRegion | EmployeeCity |
| --- | --- | --- |
| USA | WA | Redmond |
| USA | WA | Tacoma |
| USA | WA | Seattle |

```sql
--use TSQLV6
--Query to be converted

USE TSQLV6;

WITH EXCEPT_ALL
AS
(
  SELECT
    ROW_NUMBER() 
      OVER(PARTITION BY country, region, city
           ORDER     BY (SELECT 0)) AS rownum,
    country, region, city
  FROM HR.Employees

  EXCEPT

  SELECT
    ROW_NUMBER() 
      OVER(PARTITION BY country, region, city
           ORDER     BY (SELECT 0)),
    country, region, city
  FROM Sales.Customers
)
SELECT country, region, city
FROM EXCEPT_ALL;


--use Northwinds2022TSQLV7
--Converted Query

USE Northwinds2022TSQLV7;

WITH EXCEPT_ALL
AS
(
  SELECT
    ROW_NUMBER() 
      OVER(PARTITION BY EmployeeCountry, EmployeeRegion, EmployeeCity
           ORDER     BY (SELECT 0)) AS rownum,
   EmployeeCountry, EmployeeRegion, EmployeeCity
  FROM HumanResources.Employee

  EXCEPT

  SELECT
    ROW_NUMBER() 
      OVER(PARTITION BY CustomerCountry, CustomerRegion, CustomerCity
           ORDER     BY (SELECT 0)),
    CustomerCountry, CustomerRegion, CustomerCity
  FROM Sales.Customer
)
SELECT EmployeeCountry, EmployeeRegion, EmployeeCity
FROM EXCEPT_ALL;
```

## **Explanation:**

The code defines a temporary result set named EXCEPT\_ALL using a CTE. It selects data from the HumanResources.Employee table and assigns unique row numbers for each combination of EmployeeCountry, EmployeeRegion, and EmployeeCity. The keyword EXCEPT is used to subtract the result of one query from another. The second query selects data from the Sales.Customer table, assigning unique row numbers for each combination. The final result set contains unique combinations of employee locations not found in the customer locations.

## _Top 4: Two Most Recent Orders for Employees 3 and 5_

- Create a query obtaining the top two orders for employees with IDs 3 and 5, combining the results into a single dataset using the UNION ALL operator.
- Tables involved: Sales.Order

Desired output for TSQLV6 database

| empid | orderid | orderdate |
| --- | --- | --- |
| 3 | 11063 | 2022-04-30 |
| 3 | 11057 | 2022-04-29 |
| 5 | 11043 | 2022-04-22 |
| 5 | 10954 | 2022-03-17 |

Desired output for Northwinds2022TSQLV7 database

| EmployeeId | OrderId | OrderDate |
| --- | --- | --- |
| 3 | 11063 | 2016-04-30 |
| 3 | 11057 | 2016-04-29 |
| 5 | 11043 | 2016-04-22 |
| 5 | 10954 | 2016-03-17 |

```sql
--use TSQLV6
--Query to be converted

USE TSQLV6;

SELECT empid, orderid, orderdate
FROM (SELECT TOP (2) empid, orderid, orderdate
      FROM Sales.Orders
      WHERE empid = 3
      ORDER BY orderdate DESC, orderid DESC) AS D1

UNION ALL

SELECT empid, orderid, orderdate
FROM (SELECT TOP (2) empid, orderid, orderdate
      FROM Sales.Orders
      WHERE empid = 5
      ORDER BY orderdate DESC, orderid DESC) AS D2;

--use Northwinds2022TSQLV7
--Converted Query

USE Northwinds2022TSQLV7;

SELECT EmployeeId, OrderId, OrderDate
FROM (SELECT TOP (2) EmployeeId, OrderId, OrderDate
      FROM Sales.[Order]
      WHERE EmployeeId = 3
      ORDER BY OrderDate DESC, OrderId DESC) AS D1

UNION ALL

SELECT EmployeeId, OrderId, OrderDate
FROM (SELECT TOP (2) EmployeeId, OrderId, OrderDate
      FROM Sales.[Order]
      WHERE EmployeeId = 5
      ORDER BY OrderDate DESC, OrderId DESC) AS D2;
```

## **Explanation:**

This query selects columns empid, orderid, and orderdate from the Sales.Orders table. It then filters the top 2 orders for each employee using a subquery. The UNION ALL operator combines the results of two SELECT queries, including all rows, even duplicates, to ensure complete data.
