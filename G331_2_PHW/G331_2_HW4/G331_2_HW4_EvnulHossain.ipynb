## _TOP 3: EXCEPT OPERATOR_

- Create a query where we retreieve the country, region, and city from the employees table that are NOT included in the customer's table.
- Table provided: HR.Employees and Sales.Customers From TSQLV6
- Desired Output:

**TSQLV6:**

| country | region | city |
| --- | --- | --- |
| USA | WA | Redmond |
| USA | WA | Tacoma |

**Northwinds2022TSQLV7**

| EmployeesCountry | EmployeeRegion | EmployeeCity |
| --- | --- | --- |
| USA | WA | Redmond |
| USA | WA | Tacoma |

```sql
------------------------------------------------------------------
-- TSQLV6
------------------------------------------------------------------
USE TSQLV6;

SELECT country, region, city FROM HR.Employees
EXCEPT
SELECT country, region, city FROM Sales.Customers;

------------------------------------------------------------------
-- NorthwindsTSQLV7
------------------------------------------------------------------
USE Northwinds2022TSQLV7;

SELECT EmployeeCountry, EmployeeRegion, EmployeeCity FROM HumanResources.[Employee]
EXCEPT
SELECT CustomerCountry, CustomerRegion, CustomerCity FROM Sales.[Customer];
```

### **Explanation:**

In the code above, we are first _SELECT_ country, region, and city FROM our HR.Employees table. However, we use a operator called _EXCEPT_ which will return out any rows that are not present in our Sales.Customer's table. We do this by _SELECT_ the customer's country, region and city _FROM_ the Sales.Customers table. This query can be very useful for filtering out results to trackdown specific statistical needs when trying to see your Employees compared to your Customers.

## _TOP 4: Using Parenthesis_

- Create a query that returns the locations of the supplier that are NOT similar to the employee 's location BUT that are similar to the customer's location. This should include the country, region, and city of these tables.
- Tables Involved: Production.Suppliers, HR.Employees, and Sales.Customers from TSQLV6
- Desired Output:

**TSQLV6:**

| country | region | city |
| --- | --- | --- |
| Canada | Québec | Montréal |
| France | NULL | Paris |
| Germany | NULL | Berlin |

**Northwinds2022TSQLV7:**

| SupplierCountry | SupplierRegion | SupplierCity |
| --- | --- | --- |
| Canada | Québec | Montréal |
| France | NULL | Paris |
| Germany | NULL | Berlin |

```sql
------------------------------
--TSQLV6
------------------------------

USE TSQLV6;

(SELECT country, region, city FROM Production.Suppliers
 EXCEPT
 SELECT country, region, city FROM HR.Employees)
INTERSECT
SELECT country, region, city FROM Sales.Customers;

------------------------------
--TSQLV6
------------------------------

USE Northwinds2022TSQLV7;

(SELECT SupplierCountry, SupplierRegion, SupplierCity FROM Production.[Supplier]
 EXCEPT
 SELECT EmployeeCountry, EmployeeRegion, EmployeeCity FROM HumanResources.[Employee])
INTERSECT
SELECT CustomerCountry, CustomerRegion, CustomerCity FROM Sales.[Customer];
```

### **Explanation:**

The sql code above we are going to _SELECT_ the country, region and city _FROM_ Production.Suppliers which is the suppliers location. However, we are using the _EXCEPT_ operator to not have any similar rows with HR.Employees table which includes the country, region, and city rows. We then put _parenthesis_ to do this part of the execution of the query first. After that is filtered out then we use the _INTERSECT_ operator to match any similar rows the _parenthesis_ result has with Sales.Customers table which includes rows from country, region, and city columns. This is very useful because we can learn how to decrease chances of having any errors due to parenthesis having the highest precendence so it will execute whatever inside the _parenthesis_ first. Also it improves readability of the code.

## _Excercise 1 Chapter 6:_

- Explain the difference between the UNION ALL and UNION operators. In what cases are the two equivalent? When they are equivalent, which one should you use?

  

### **Solution:**

 The difference between _UNION ALL_ and _UNION_ operators is how _UNION ALL_ operator will combine two tables regardless of rows that have the same exact values which is basically duplicated. In the _UNION_ operator, it will filter and remove any rows that have the same exact values overall eliminating any duplicates. A case where both of these operators are equivalent is when both tables that are being combined already does NOT have any rows with identical values from each other tables. Between choosing UNION ALL and UNION, choosing UNION would be the perferred way due to most queries in the real world would not want any duplicates and it would overall avoid more work being done to check unless the query does not care about duplicates coming up in the results.
