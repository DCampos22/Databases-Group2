## Top 1 (Medium)

\- Create a query that shows the credit card information of all businesses (id), used credit cards by using their IDs, and their most recent order which includes the orderID and OrderDate from SalesOrderHeader table. 

I chose this problem due to how we can see how left join can be used to compare two tables for certain columns.

Tables Involved: Sales.CreditCard, Sales.PersonCreditCard, Sales.PersonCreditCard 

_Standard View:_ 



_Column View:_



_Key View:_

  

Columns: 
| Table Names                | Columns Selected          |
|----------------------------|---------------------------|
| Sales.CreditCard           | CreditCardID              |
| Sales.PersonCreditCard     | BusinessEntityID, CreditCardID |
| Sales.SalesOrderHeader     | CreditCardID, SalesOrderID, OrderDate |

```sql
USE AdventureWorks2017;

SELECT DISTINCT
       pc.BusinessEntityID,
	   pc.CreditCardID,
	   cc.CardType,
	   cc.CardNumber,
	   cc.ExpMonth,
	   cc.ExpYear,
       soh.OrderId,
	   soh.OrderDate AS MostRecentOrderDate
FROM Sales.CreditCard AS cc
INNER JOIN Sales.PersonCreditCard AS pc
    ON cc.CreditCardID = pc.CreditCardID
LEFT JOIN (
    SELECT CreditCardID, MAX(OrderDate) AS OrderDate,
    SalesOrderID AS OrderId
    FROM Sales.SalesOrderHeader
    GROUP BY CreditCardID, SalesOrderID
) AS soh ON cc.CreditCardID = soh.CreditCardID
ORDER BY pc.CreditCardID
FOR JSON PATH, ROOT('Medium Top #1'), INCLUDE_NULL_VALUES;
```

```sql
use AdventureWorks2017;

WITH CustomerOrderSummary AS (
    SELECT
        Customer.CustomerID,
        COUNT(DISTINCT OrderHeader.SalesOrderID) AS NumOrders,
        AVG(OrderDetails.UnitPrice * OrderDetails.OrderQty) AS AvgOrderTotal
    FROM
        Sales.Customer AS Customer
    JOIN
        Sales.SalesOrderHeader AS OrderHeader ON Customer.CustomerID = OrderHeader.CustomerID
    JOIN
        Sales.SalesOrderDetail AS OrderDetails ON OrderHeader.SalesOrderID = OrderDetails.SalesOrderID
    GROUP BY
        Customer.CustomerID
)

SELECT
    Customer.CustomerID,
    Person.FirstName,
    Person.LastName,
    OrderSummary.NumOrders,
    OrderSummary.AvgOrderTotal
FROM
    Sales.Customer AS Customer
JOIN
    Person.Person AS Person ON Customer.PersonID = Person.BusinessEntityID
JOIN
    CustomerOrderSummary AS OrderSummary ON Customer.CustomerID = OrderSummary.CustomerID
ORDER BY
    Customer.CustomerID
    FOR JSON PATH, ROOT('Complex Top #2'), INCLUDE_NULL_VALUES;
```

## Top 3 (Medium)

Create a query that retrieves customer's order total along with their id and name which will show which customer made the highest purchase total in descending order.

Another example of CTE which is why I appreciate this query because it simply shows how we can obtain totalprice in a CTE.

Tables Involved: Sales.Customers, Sales.Orders, Sales.OrderLines

_Standard View:_



Columns View:



KEY View:





| Table Names                | Columns Selected                           |
|----------------------------|--------------------------------------------|
| Sales.Orders               | CustomerID                                 |
| Sales.OrderLines           | OrderID, Quantity, UnitPrice               |
| Sales.Customers            | CustomerID, CustomerName                   |


```sql
use WideWorldImporters;

WITH TotalSales AS (
    SELECT
        O.CustomerID,
        SUM(OD.Quantity * OD.UnitPrice) AS TotalPrice
    FROM
        Sales.Orders AS O
    JOIN
        Sales.OrderLines AS OD ON O.OrderID = OD.OrderID
    GROUP BY
        O.CustomerID
)

SELECT
    C.CustomerID,
    C.CustomerName,
    SUM(TS.TotalPrice) AS TotalSalesAmount
FROM
    Sales.Customers AS C
INNER JOIN
    TotalSales AS TS ON C.CustomerID = TS.CustomerID
GROUP BY
    C.CustomerID,
    C.CustomerName
ORDER BY
    TotalSalesAmount DESC
    
```

## Worst 1( Complex )

\- Write a query where using a CTE, we find the stock item's age in days, the total price of the item in descending order, stock item information (key, item name, and the supplier associated with it. However, if the total price of a item is null replace it with 0.

Tables Involved: Fact.\[Purchase\], Dimension.\[Stock Item\], Dimension.\[Stock Item\]

_Standard View:_



Column View:



Key View:

  

  

| Table Names | Columns Selected |
| --- | --- |
| Fact.Purchase | StockItemKey, SupplierKey |
| Dimension.StockItem | StockItemKey, StockItem, RecommendedRetailPrice, ValidFrom, ValidTo |
| Dimension.Supplier | SupplierKey, Supplier |

```sql
USE WideWorldImportersDW;

WITH ProductSales AS (
    SELECT
        s.[Stock Item Key] AS StockItemKey,
        i.[Stock Item] AS StockItemName,
        ISNULL(SUM(i.[Recommended Retail Price]), 0) AS ItemPrice,
         DATEDIFF(DAY, i.[Valid From], i.[Valid To]) AS ProductAgeInDays,
        sup.[Supplier]  AS SupplierName
    FROM
        Fact.[Purchase] AS s
    JOIN
        Dimension.[Stock Item] AS i ON s.[Stock Item Key] = i.[Stock Item Key]
    JOIN
        Dimension.[Supplier] AS sup ON s.[Supplier Key] = sup.[Supplier Key]
    GROUP BY
        s.[Stock Item Key],
        i.[Stock Item],
        sup.[Supplier],
       DATEDIFF(DAY, i.[Valid From], i.[Valid To])
)

SELECT
    StockItemKey,
    StockItemName,
    ItemPrice,
    ProductAgeInDays,
    SupplierName
FROM
    ProductSales
WHERE
    ProductAgeInDays IS NOT NULL
ORDER BY
    ItemPrice DESC
    FOR JSON PATH, ROOT('Complex Worst #1'), INCLUDE_NULL_VALUES;
```

## Worst 3 (Complex)

Create a query were we need a function that will take the orderID and see if the type of order verifies for warrenty vip 1 which equals PO or vip 2 which equals SO or N/A which equals OM. We will also return product id, unit price, quantity, discount percentage, and the orderdate.

Tables Involved: Sales.Order, Sales.OrderDetail, Production.Product

  

Standard View:

  

  

Column View:

  

Key View:



```sql
use Northwinds2022TSQLV7;


GO
-- Create a function to determine membership level based on order type
CREATE FUNCTION dbo.GetMembershipLevel
(
    @OrderID INT
)
RETURNS NVARCHAR(50)
AS
BEGIN
    DECLARE @OrderType VARCHAR(50);

    -- Get the type of order based on OrderID
    SET @OrderType = Sales.TypesOfOrders(@OrderID);

    -- Determine membership level based on order type
    IF @OrderType LIKE 'PO%'
        RETURN 'VIP1 Membership';
    ELSE IF @OrderType LIKE 'SO%'
        RETURN 'VIP2 Membership';
    ELSE IF @OrderType LIKE 'OM%'
        RETURN 'N/A';
   
    RETURN 'Unknown'; -- Return 'Unknown' if the order type doesn't match any criteria
END;
GO

SELECT 
    Sales.TypesOfOrders(o.OrderID) AS [Type Order with ID],
    dbo.GetMembershipLevel(SUBSTRING(Sales.TypesOfOrders(o.OrderID), 4, LEN(Sales.TypesOfOrders(o.OrderID)) - 3)) AS [Warrenty],
    od.ProductID,
    p.ProductName,
    od.UnitPrice,
    od.Quantity,
    od.DiscountPercentage,
    -- Discounted Price per unit
    o.OrderDate AS [Order Date]    
FROM 
    Sales.[Order] AS o
JOIN 
    Sales.[OrderDetail] AS od ON o.OrderID = od.OrderID
JOIN 
    Production.[Product] AS p ON od.ProductID = p.ProductID
ORDER BY  
    dbo.GetMembershipLevel(SUBSTRING(Sales.TypesOfOrders(o.OrderID), 4, LEN(Sales.TypesOfOrders(o.OrderID)) - 3))
   
```

Create a query that returns each customer's information including how many orders they placed , and their average total they spent on their orders.

Tables Involved: Sales.Customer, Sales.SalesOrderHeader, Sales.SalesOrderDetail

```sql
use AdventureWorks2017;

WITH CustomerOrderSummary AS (
    SELECT
        Customer.CustomerID,
        COUNT(DISTINCT OrderHeader.SalesOrderID) AS NumOrders,
        AVG(OrderDetails.UnitPrice * OrderDetails.OrderQty) AS AvgOrderTotal
    FROM
        Sales.Customer AS Customer
    JOIN
        Sales.SalesOrderHeader AS OrderHeader ON Customer.CustomerID = OrderHeader.CustomerID
    JOIN
        Sales.SalesOrderDetail AS OrderDetails ON OrderHeader.SalesOrderID = OrderDetails.SalesOrderID
    GROUP BY
        Customer.CustomerID
)

SELECT
    Customer.CustomerID,
    Person.FirstName,
    Person.LastName,
    OrderSummary.NumOrders,
    OrderSummary.AvgOrderTotal
FROM
    Sales.Customer AS Customer
JOIN
    Person.Person AS Person ON Customer.PersonID = Person.BusinessEntityID
JOIN
    CustomerOrderSummary AS OrderSummary ON Customer.CustomerID = OrderSummary.CustomerID
ORDER BY
    Customer.CustomerID;
    --FOR JSON PATH, ROOT('Complex Q9'), INCLUDE_NULL_VALUES;
```

## Q11 (complex)

\- Write a query where using a CTE, we find the stock item's age in days, the total price of the item in descending order, stock item information (key, item name, and the supplier associated with it. However, if the total price of a item is null replace it with 0.

Tables Involved: Fact.\[Purchase\], Dimension.\[Stock Item\], Dimension.\[Stock Item\]

```sql
USE WideWorldImportersDW;

WITH ProductSales AS (
    SELECT
        s.[Stock Item Key] AS StockItemKey,
        i.[Stock Item] AS StockItemName,
        ISNULL(SUM(i.[Recommended Retail Price]), 0) AS ItemPrice,
         DATEDIFF(DAY, i.[Valid From], i.[Valid To]) AS ProductAgeInDays,
        sup.[Supplier]  AS SupplierName
    FROM
        Fact.[Purchase] AS s
    JOIN
        Dimension.[Stock Item] AS i ON s.[Stock Item Key] = i.[Stock Item Key]
    JOIN
        Dimension.[Supplier] AS sup ON s.[Supplier Key] = sup.[Supplier Key]
    GROUP BY
        s.[Stock Item Key],
        i.[Stock Item],
        sup.[Supplier],
       DATEDIFF(DAY, i.[Valid From], i.[Valid To])
)

SELECT
    StockItemKey,
    StockItemName,
    ItemPrice,
    ProductAgeInDays,
    SupplierName
FROM
    ProductSales
WHERE
    ProductAgeInDays IS NOT NULL
ORDER BY
    ItemPrice DESC;
    --FOR JSON PATH, ROOT('Complex Q11'), INCLUDE_NULL_VALUES;
```

\- Create a query where we need a CTE table that shows the Cost Change of each product including its id and its name. If the cost increased write increased on the value and if the cost decreased write decrease as the value for the column.

Tables Involved: Purchasing.Vendor, Production.[Product]

```sql
USE AdventureWorks2017;

WITH CostChange AS (
    SELECT 
        ph.ProductID,
        ph.StandardCost,
        LAG(ph.StandardCost) 
           OVER (
            PARTITION BY p.ProductID 
             ORDER BY ph.StartDate
             ) AS PreviousCost,
        p.Name AS ProductName
    FROM 
        Production.[ProductCostHistory] AS ph
    JOIN
        Production.[Product] AS p 
          ON p.ProductID = ph.ProductID
)
SELECT 
    ProductID,
    ProductName,
    StandardCost,
    PreviousCost,
    CASE 
        WHEN StandardCost > PreviousCost THEN 'Increase'
        WHEN StandardCost < PreviousCost THEN 'Decline'
        ELSE 'No Change'
    END AS ChangeType
FROM 
    CostChange;
        --FOR JSON PATH, ROOT('Complex Q16'), INCLUDE_NULL_VALUES;
```

## Q18 (medium)

\- Create a query where we create a function that returns table that retrieves the total sales from each customer from the AdventureWorksDW2017 database. When we call the function in the query we will make sure the customer's yearly income is over 60000 dollars.

Tables Involved: dbo.DimCustomer, dbo.FactInternetSales

```sql
USE AdventureWorksDW2017;
GO

CREATE FUNCTION dbo.ufnGetCustomerSales(@CustomerKey INT)
RETURNS TABLE
AS
RETURN 
(
    SELECT
         c.CustomerKey as CustomerID, 
         SUM(f.SalesAmount) AS TotalSales
        
    FROM 
        dbo.FactInternetSales AS f
    JOIN 
        dbo.DimCustomer AS c 
        ON 
        f.CustomerKey = c.CustomerKey
    WHERE 
        c.CustomerKey = @CustomerKey    
    GROUP BY c.CustomerKey
);

GO 

SELECT 
    c.CustomerKey,
    s.TotalSales,
    C.YearlyIncome
FROM 
    dbo.DimCustomer AS c
CROSS APPLY 
    dbo.ufnGetCustomerSales(c.CustomerKey) AS s
WHERE
    c.YearlyIncome >=60000.00
ORDER BY
    c.YearlyIncome;
            --FOR JSON PATH, ROOT('Medium Q18'), INCLUDE_NULL_VALUES;
```

## Q19 (complex)

Find the calendar year quarter of all order dates from Sales.Order table, the ProductID and the ProductName from Production.Product by creating a function.

Tables Invovled: Sales,Order, Sales.OrderDetails, Production.Product

```sql
USE Northwinds2022TSQLV7;

GO
CREATE OR ALTER FUNCTION Sales.udf_CalendarYearQuarter
(
    @OrderDate DATE
)
RETURNS CHAR(100)
BEGIN
    DECLARE @QUARTER CHAR(100);

    IF MONTH(@OrderDate) >= 1 AND MONTH(@OrderDate) <= 3
        SET @QUARTER = 'Q1';
    ELSE IF MONTH(@OrderDate) >= 4 AND MONTH(@OrderDate) <= 6
        SET @QUARTER = 'Q2';
    ELSE IF MONTH(@OrderDate) >= 7 AND MONTH(@OrderDate) <= 9
        SET @QUARTER = 'Q3';
    ELSE IF MONTH(@OrderDate) >= 10 AND MONTH(@OrderDate) <= 12
        SET @QUARTER = 'Q4';
    ELSE
        SET @QUARTER = 'N/A';

    RETURN @QUARTER;
END;
GO

SELECT o.OrderId,
       od.ProductId,
       pp.ProductName,
	   YEAR(o.OrderDate) AS orderyear,
       Sales.udf_CalendarYearQuarter(o.OrderDate) AS [Calendar Quarterly]   
FROM Sales.[Order] AS o
    INNER JOIN Sales.OrderDetail AS od
        ON od.OrderId = o.OrderId
    INNER JOIN Production.Product AS pp
        ON od.ProductId = pp.ProductId
GROUP BY o.OrderId,
	     o.OrderDate,
         od.ProductId,
         pp.ProductName
ORDER BY o.OrderId;
    --FOR JSON PATH, ROOT('Complex Q19'), INCLUDE_NULL_VALUES;
```

## Q20 (Medium)

\- Create a query using the datekey from FactCurrencyRate, the french version of the date from DimDate, and rounded average rate of currency rate from FactCurrencyRate.

Tables Involved: dbo.FactCurrencyRate, fcr.DateKey

```sql
USE AdventureWorksDW2017;
SELECT fcr.DateKey,
       CONCAT(dd.FrenchMonthName, ' ', dd.FrenchDayNameOfWeek, ' ', dd.CalendarYear) AS spanishdate,
       CONCAT(ROUND(fcr.AverageRate * 100, 0), '%') AS AverageRate
FROM dbo.DimDate AS dd
    INNER JOIN dbo.FactCurrencyRate AS fcr
        ON dd.DateKey = fcr.DateKey
GROUP BY fcr.DateKey,
         CONCAT(dd.FrenchMonthName, ' ', dd.FrenchDayNameOfWeek, ' ', dd.CalendarYear),
                fcr.AverageRate * 100 
ORDER BY fcr.DateKey;
--FOR JSON PATH, ROOT('Medium Q15'), INCLUDE_NULL_VALUES;
```
